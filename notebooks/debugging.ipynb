```python
import xmltodict

def read_xml(path):
    

    with open(path) as file:
        file_data = file.read()
        dict_data = xmltodict.parse(file_data)

    return dict_data
        
```

```python
import sys 

sys.path.append('../')

import matplotlib.pyplot as plt 
import cv2 

from datasets.pascal_voc import PASCALVOCDataset
from pipelines.transforms import Rescale, RandomCrop
from torchvision import transforms

root_dir = '/home/user/UOS-SSaS Dropbox/05. Data/00. Benchmarks/02. PASCAL VOC/VOCdevkit/VOC2012'

composed = transforms.Compose([Rescale(512),
                               RandomCrop(480) ])

dataset = PASCALVOCDataset(root_dir=root_dir, transform=composed)

```

```python
from random import randint 

idx = randint(0, len(dataset))

sample = dataset[idx]
img, bboxes = sample['image'], sample['bbox']

for bbox in bboxes:
    start_point = bbox[0], bbox[1]
    end_point = bbox[2], bbox[3]

    img = cv2.rectangle(img, start_point, end_point, (255, 0, 0), 2)

plt.imshow(img)
```
